```python
import csv
import json
import pandas as pd

def convert_csv(text):
    json_data = text

    # Chuyển đổi JSON thành một đối tượng Python
    data = json.loads(json_data)

    # Tạo DataFrame từ dữ liệu JSON
    df = pd.DataFrame(data['days'])

    
    print("Chuyển đổi thành công và lưu thành DataFrame!")
    return df
```

```python
# convert_csv(response_1.text)
# convert_csv(response2.text)
meta = pd.concat([convert_csv(response_1.text), convert_csv(response_2.text)]).reset_index()
meta.head(5)
```

```python
### Dữ liệu 0 nghĩa là ngày mai sẽ không mưa, 1 là ngày mai sẽ mưa

for i in meta.index[: len(meta.index)-1]:
    if meta.precip[i+1] > 0:
        meta.precip[i] = 1
    else:
        meta.precip[i] = 0
```

```python
meta.drop(index=[len(meta)-1], axis=1, inplace=True)
```
