```python
import requests
from bs4 import BeautifulSoup
import time
import os
html_text = requests.get('https://sixdo.vn/uu-dai-p36?gclid=Cj0KCQjwuLShBhC_ARIsAFod4fK-YkZnWP7KwARUXXCN04I78HP9Dy2W7Miy3uPMsar3MsglrVpV1PIaAnduEALw_wcB').text
soup = BeautifulSoup(html_text, 'lxml')
soup
```

```python
list_problem = soup.find_all('ul', id='childMenu_0')
list_problem
```

```python
products = []
```

```python

# Tìm tất cả các phần tử a có class là sub_item 1 trong từng phần tử ul
danh_mucs = []
for ul in list_problem:
    a_tags = ul.find_all('a', class_='sub_item 1')
    for paper in a_tags:
            path = paper['href']
            category = paper.text 
            
            html_p = requests.get(path).text
            paper = BeautifulSoup(html_p, 'lxml')
            
            while True:
                box = paper.find_all('div', class_='row c-box_list')[0]
                list_iterm = box.find_all('div')
                cnt = 0
                # break
    
            
                for iterm in list_iterm:
                    # Tìm phần tử div chứa thông tin sản phẩm
                    # print(iterm)
                    try:
                        link_img = iterm.find('a').img['src']
                        # print(link_img)
                    except:
                        continue
                    namesp = iterm.find('h3', class_="c-name").text.strip()
                    price = int(iterm.find('span', class_="c-real_price").text.strip().replace('.','').replace('đ',''))
                    print(price)
                    product = {
                        'name': namesp,
                        'price': price,
                        'category': category
                    }
                    products.append(product)
                    name_img = namesp.replace(' ', '_')
                    response = requests.get(link_img)
                    with open(f'images/{name_img}.jpg', 'wb') as f:
                        f.write(response.content)
                    cnt+=1
                print(f'{category} = {cnt}')
                # print("111")
                try:
                    next_paper = paper.find('a', rel="nofollow", class_="next-page")['href']
                    
                except:
                    break
                # paper = requests.get(f'https://sixdo.vn/{next_paper}').text
                html_p = requests.get(f'https://sixdo.vn/{next_paper}').text
                paper = BeautifulSoup(html_p, 'lxml')
                print(next_paper)
            # break
# In ra nội dung của các phần tử a
# for danh_muc in danh_mucs:
#     print(danh_muc.text)
```
